```python
# import the needed packages
import cv2
import os
import pytesseract

# Read image - directly specify the path
# image_path = args["image"] # Removed argparse related lines
image_path = "/content/OIP.png" # Using the path to image file
image = cv2.imread(image_path) # read the image

# if no image found
if image is None:
    raise FileNotFoundError("Image not found or invalid path")

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Preprocessing
# Using a default pre-processor for simplicity in notebook context
# if args["pre_processor"] == "thresh": # Removed argparse related lines
#     gray = cv2.threshold(
#         gray, 0, 255,
#         cv2.THRESH_BINARY | cv2.THRESH_OTSU
#     )[1]
# elif args["pre_processor"] == "blur":
#     gray = cv2.medianBlur(gray, 3)

# Default to 'thresh' preprocessing
gray = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

# OCR (no temp file needed)
custom_config = r"--oem 3 --psm 6"

# converting detected text in image to string
text = pytesseract.image_to_string(gray, config=custom_config)

# show's output
print("Detected Text:")
print(text)

# Optional: display images (comment out if running headless)
# In a Colab environment, cv2.imshow requires a workaround (e.g., using google.colab.patches.cv2_imshow)
# For simplicity, we'll use matplotlib for display if needed, or remove display for just text extraction
# To avoid errors with cv2.imshow not working directly in Colab without setup:
# Removed cv2.imshow, cv2.waitKey, cv2.destroyAllWindows for Colab compatibility.
# If display is needed, consider using matplotlib or google.colab.patches.cv2_imshow
```
